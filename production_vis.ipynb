```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', 200)
import datetime
import seaborn as sns
import matplotlib.ticker as mtick
```

```python
path = 'Data/Errorbars/final.xlsx'
best_zoned_case = ['e11_v00139', 'Best zoned']
best_nonzoned_case = ['e9_v00147', 'Best non-zoned']

history_data = 'History data'
title = 'Comparison of history matching'
fontsize = 10
scatter_point_size = 8

xmin = datetime.datetime(2011, 1, 1)
xmax = 	datetime.datetime(2019, 3, 12)
xlabel = 'Year'

ylabel_oil = r'$Q_{oil}$'+' STB/day'
ylabel_wat =  r'$Q_{water}$'+', STB/day'
ylabel_p = 'BHP, psia'
```

Well

```python
well = '4A' # options: 2, 4A, 5, 10

i1 = 'WOPR_' + well
i2 = 'WWPR_' + well
i3 = 'BHP_' + well
title = 'History matching of ' + well + ' well'
title_WOPR = 'WOPR'# of ' + well + ' well'
title_WWPR = 'WWPR'# of ' + well + ' well'
title_BHP = 'BHP'# of ' + well + ' well'

ymin_oil = 0
ymax_oil = 9000
ystep_oil = 2000
ymin_wat = 0
ymax_wat = 1500
ystep_wat = 200
ymin_p = 0
ymax_p = 2800
ystep_p = 600

# oil
if well == '2':
    ymax_oil = 8500
    ystep_oil = 2000
if well == '4A':
    ymax_oil = 10000
    ystep_oil = 2000
if well == '5':
    ymax_oil = 10000
    ystep_oil = 2000
if well == '10':
    ymax_oil = 9000
    ystep_oil = 2000

# water
if well == '2':
    ymax_wat = 2500
    ystep_wat = 500
if well == '5':
    ymax_wat = 2500
    ystep_wat = 500
if well == '10':
    ymax_wat = 1000
    ystep_wat = 250
if well == '4A':
    ymax_wat = 250
    ystep_wat = 50

# pressure
if well == '2':
    ymax_p = 2600
    ystep_p = 500
if well == '4A':
    ymax_p = 2500
    ystep_p = 500
if well == '5':
    ymax_p = 3000
    ystep_p = 500
if well == '10':
    ymax_p = 3000
    ystep_p = 500


variable = 'profiles_of_' + well + '_well'
plt.figure(figsize = (8, 8))
plt.suptitle(title, fontsize = fontsize)

# WOPR
plt.subplot(3, 1, 1)
plt.title(title_WOPR, fontsize = fontsize)
data = pd.read_excel(path, sheet_name = i1)
plt.scatter(x = data[''].values, y = data['. '].values, label = history_data, s = 5)
plt.plot(data[''], data[best_zoned_case[0]], label = best_zoned_case[1], color = 'red')
plt.plot(data[''], data[best_nonzoned_case[0]], label = best_nonzoned_case[1], color = 'green')
plt.grid(True)
plt.xlabel(xlabel, fontsize = fontsize)
plt.ylabel(ylabel_wat, fontsize = fontsize)
plt.xlim(xmin, xmax)
plt.ylim(ymin_oil, ymax_oil)
legend = plt.legend(loc = 'lower left', ncol = 1, borderaxespad = 1, fontsize = fontsize, frameon = True, fancybox = True, framealpha = 0.5)
legend.get_frame()
plt.yticks(np.arange(ymin_oil, ymax_oil, ystep_oil))

# WWPR
plt.subplot(3, 1, 2)
plt.title(title_WWPR, fontsize = fontsize)
data = pd.read_excel(path, sheet_name = i2)
plt.scatter(x = data[''].values, y = data['. '].values, label = history_data, s = 5)
plt.plot(data[''], data[best_zoned_case[0]], label = best_zoned_case[1], color = 'red')
plt.plot(data[''], data[best_nonzoned_case[0]], label = best_nonzoned_case[1], color = 'green')
plt.grid(True)
plt.xlabel(xlabel, fontsize = fontsize)
plt.ylabel(ylabel_wat, fontsize = fontsize)
plt.xlim(xmin, xmax)
plt.ylim(ymin_wat, ymax_wat)
legend = plt.legend(loc = 'upper left', ncol = 1, borderaxespad = 1, fontsize = fontsize, frameon = True, fancybox = True, framealpha = 0.5)
legend.get_frame()
plt.yticks(np.arange(ymin_wat, ymax_wat, ystep_wat))

# BHP
plt.subplot(3, 1, 3)
plt.title(title_BHP, fontsize = fontsize)
data = pd.read_excel(path, sheet_name = i3)
plt.scatter(x = data[''].values, y = data['. '].values, label = history_data, s = 5)
plt.plot(data[''], data[best_zoned_case[0]], label = best_zoned_case[1], color = 'red')
plt.plot(data[''], data[best_nonzoned_case[0]], label = best_nonzoned_case[1], color = 'green')
plt.grid(True)
plt.xlabel(xlabel, fontsize = fontsize)
plt.ylabel(ylabel_p, fontsize = fontsize)
plt.xlim(xmin, xmax)
plt.ylim(ymin_p, ymax_p)
legend = plt.legend(loc = 'lower left', ncol = 1, borderaxespad = 1, fontsize = fontsize, frameon = True, fancybox = True, framealpha = 0.5)
legend.get_frame()
plt.yticks(np.arange(ymin_p, ymax_p, ystep_p))

plt.tight_layout()
plt.subplots_adjust(top = 0.9);
plt.savefig('/Users/antonvoskresenskii/Documents/Krapivninskoe/WL/WATT_FIELD/Figures/' + variable, dpi = 200)
```
